# Step 1: Load Dataset

Machine Learning Repository https://archive.ics.uci.edu/dataset/109/wine

```python
import pandas as pd

# URL of the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"


# Load the dataset into DataFrame
wine_df = pd.read_csv(url, header=None, index_col=False, names=[
                'Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
                'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
                'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                'Proline'])

# Display wine_data DF and check that all features are listed
display(wine_df)
```

Confirming that all features (13) and instances (178) as listed on the website have been loaded.

# Explore the dataset
*note that class is a categorical variable, but has Dtype as int below

```python
# Show info on wine_df
print("Info on wine_df:")
wine_df.info()
print()


# Show description of wine_df
print("Description of wine_df:")
wine_df.describe()
```

# Checking for missing values in dataset before splitting

The dataset on UCI ML repository list zero missing values. Double checking below and confirming there are zero missing values.

```python
# Check for missing values in wine_df
missing_values = wine_df.isnull().sum()

# Print missing values in wine_df
print("Missing Values:")
print(missing_values)
```

# Step 2: Divide into a training set and a testing set

To create a balanced split StratifiedShuffleSplit will be utilized below

```python
# Utilize scikitlearn for a stratify split
from sklearn.model_selection import StratifiedShuffleSplit

# Separate the features (X) and the target variable (y) aka Class
X = wine_df.drop('Class', axis=1)
y = wine_df['Class']

# Create StratifiedShuffleSplit with 1 split, 20% test size, 80% training and random state 42
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
# Split data
for train_index, test_index in split.split(X, y):
   X_train = X.iloc[train_index]
   X_test = X.iloc[test_index]
   y_train = y.iloc[train_index]
   y_test = y.iloc[test_index]

# Check to confirm successful 80 : 20 split
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

Confirming the split was successful. The training set has 80% of the original 178 instances, while the testing set has 20% of the original instances. 13 in the X-train and X_test show the number of columns/features.

# Step 6: Apply transformations to data

Apply the below transformations to 2 features (Alcohol and Flavanoids) from the dataset
*  Squaring (**2)
*  Cubing (**3)
*  np.log
*  np.exp

Plot histograms and scatter matrices to explore the resultant data

# Transformations (squaring, cubbing, np.log and np.exp) applied to Alcohol feature in training set and represented as 4 histograms

```python
# No transformation applied to Alcohol feature shown as histogram
print("Original Alcohol Histogram:")
plt.hist(X_train['Alcohol'].values.reshape(-1, 1), bins=10)
plt.show()
print()

# Transformation squaring applied to Alcohol feature shown as histogram
print("Alcohol Squaring Histogram:")
plt.hist(X_train['Alcohol'].values.reshape(-1, 1)**2, bins=10)
plt.show()
print()

# Transformation cubing applied to Alcohol feature shown as histogram
print("Alcohol Cubing Histogram:")
plt.hist(X_train['Alcohol'].values.reshape(-1, 1)**3, bins=10)
plt.show()
print()

# Transformation np.log applied to Alcohol feature shown as histogram
print("Alcohol np.log Histogram:")
plt.hist(np.log(X_train['Alcohol'].values.reshape(-1, 1)), bins=10)
plt.show()
print()

# Transformation np.exp applied to Alcohol feature shown as histogram
print("Alcohol np.exp Histogram:")
plt.hist(np.exp(X_train['Alcohol'].values.reshape(-1, 1)), bins=10)
plt.show()
print()
```

```python
# No transformation applied to Flavanoids feature shown as histogram
print("Original Flavanoids Histogram:")
plt.hist(X_train['Flavanoids'].values.reshape(-1, 1), bins=10)
plt.show()
print()

# Transformation squaring applied to Flavanoids feature shown as histogram
print("Flavanoids Squaring Histogram:")
plt.hist(X_train['Flavanoids'].values.reshape(-1, 1)**2, bins=10)
plt.show()
print()

# Transformation cubing applied to Flavanoids feature shown as histogram
print("Flavanoids Cubing Histogram:")
plt.hist(X_train['Flavanoids'].values.reshape(-1, 1)**3, bins=10)
plt.show()
print()

# Transformation np.log applied to Flavanoids feature shown as histogram
print("Flavanoids np.log Histogram:")
plt.hist(np.log(X_train['Flavanoids'].values.reshape(-1, 1)), bins=10)
plt.show()
print()

# Transformation np.exp applied to Flavanoids feature shown as histogram
print("Flavanoids np.exp Histogram:")
plt.hist(np.exp(X_train['Flavanoids'].values.reshape(-1, 1)), bins=10)
plt.show()
print()
```
