The Abalone dataset used was found on the UC Irvine Machine Learning Repository.

[https://archive.ics.uci.edu/dataset/1/abalone](https://)


Features of this data set include sex, length, diameter, height, whole weight, shucked weight, viscera weight and shell weight.

The target (dependent variable) is the number of rings found in the abalone shell. The amount of rings in a shell denotes the age of the abalone.



# Step 1: Find and download the dataset


```python
import pandas as pd


# URL of the Abalone dataset on the UCI Machine Learning Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"


# Read the dataset into a pandas DataFrame
abalone_df = pd.read_csv(url, header=None, index_col=False,
    names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight',
           'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings'])

# Display the abalone_df DataFrame
display(abalone_df)
```

```python
# Check for any missing values in abalone_df
missing_values = abalone_df.isnull().sum()

# Print missing values in abalone_df
print(missing_values)
```
